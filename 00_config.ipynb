```python
# ================================================================
# FHIR Pipeline Configuration
# Central config — all notebooks import this
# ================================================================

CONFIG = {
    # API
    "fhir_base_url"  : "https://hapi.fhir.org/baseR4",
    "page_size"      : 50,
    "page_cap"       : 5,
    "ingest_days"    : 3,
    "timeout_seconds": 60,

    # Resources to ingest (order matters for pipeline)
    "resources": ["Patient", "Encounter", "Observation", "Condition"],

    # Lakehouse
    "lakehouse_name" : "fhir_lakehouse",
    "raw_base_path"  : "Files/raw",

    # Schema names
    "bronze_schema"  : "bronze",
    "silver_schema"  : "silver",
    "gold_schema"    : "gold",

    # Naming conventions
    "table_prefix"   : "tbl_",
    "view_prefix"    : "vw_",

    # SCD2
    "high_date"      : "9999-12-31",
    "business_key"   : "resource_id",
}

# ── Naming helpers ───────────────────────────────────────────────
def bronze_table(resource: str) -> str:
    """e.g. bronze.tbl_patient"""
    return f"{CONFIG['bronze_schema']}.{CONFIG['table_prefix']}{resource.lower()}"

def silver_table(resource: str) -> str:
    """e.g. silver.tbl_patient"""
    return f"{CONFIG['silver_schema']}.{CONFIG['table_prefix']}{resource.lower()}"

def gold_view(resource: str) -> str:
    """e.g. gold.vw_dim_patient"""
    suffix = GOLD_MAP[resource.lower()]
    return f"{CONFIG['gold_schema']}.{CONFIG['view_prefix']}{suffix}"

def raw_path(resource: str, date_str: str) -> str:
    """e.g. Files/raw/patient/2026-06-16/response.json"""
    return f"{CONFIG['raw_base_path']}/{resource.lower()}/{date_str}/response.json"

# ── Gold view name mapping ───────────────────────────────────────
GOLD_MAP = {
    "patient"     : "dim_patient",
    "encounter"   : "fact_encounter",
    "observation" : "fact_observation",
    "condition"   : "fact_condition",
}

# ── Print summary ────────────────────────────────────────────────
print("=" * 50)
print("FHIR Pipeline Config loaded")
print("=" * 50)
print(f"  FHIR URL   : {CONFIG['fhir_base_url']}")
print(f"  Resources  : {CONFIG['resources']}")
print(f"  Ingest days: {CONFIG['ingest_days']}")
print(f"  Schemas    : {CONFIG['bronze_schema']} | {CONFIG['silver_schema']} | {CONFIG['gold_schema']}")
print(f"  Naming     : tables → tbl_  | views → vw_")
print("\nTable names:")
for r in CONFIG["resources"]:
    print(f"  {bronze_table(r):35s} → {silver_table(r):35s} → {gold_view(r)}")
```
